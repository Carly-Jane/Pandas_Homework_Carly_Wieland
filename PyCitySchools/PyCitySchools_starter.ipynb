```python
# Dependencies and Setup
import pandas as pd
pd.options.mode.chained_assignment = None
default='warn'
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
#school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools
total_number_of_schools = len(school_data_complete_reformatted["School Name"].unique())

# Calculate the total number of students
total_number_of_students = len(school_data_complete_reformatted["Student Name"])

# Calculate the total budget
total_budget = sum(school_data["budget"])

# Calculate the average math score 
average_math_score = (sum(school_data_complete_reformatted["Math Score"])/total_number_of_students)

# Calculate the average reading score
average_reading_score = (sum(school_data_complete_reformatted["Reading Score"])/total_number_of_students)
```

```python
# Calculate the percentage of students with a passing math score (70 or greater)
number_of_students_passing_math = len(school_data_complete_reformatted.loc[school_data_complete_reformatted["Math Score"] >= 70])
percentage_of_students_passing_math = ((100/total_number_of_students)*number_of_students_passing_math)

# Creating additional dataframe for students with a passing math score (70 or greater)
number_of_students_passing_math_df = school_data_complete_reformatted.loc[school_data_complete_reformatted["Math Score"] >= 70, ["Student Name", "Student ID", "Gender", "Grade", "Reading Score", "Math Score", "School Name", "School ID", "School Type", "School Size", "School Budget"]]

# Calculate the percentage of students with a passing reading score (70 or greater)
number_of_students_passing_reading = len(school_data_complete_reformatted.loc[school_data_complete_reformatted["Reading Score"] >= 70])
percentage_of_students_passing_reading = ((100/total_number_of_students)*number_of_students_passing_reading)

# Creating additional dataframe for students with a passing math score (70 or greater)
number_of_students_passing_reading_df = school_data_complete_reformatted.loc[school_data_complete_reformatted["Reading Score"] >= 70, ["Student Name", "Student ID", "Gender", "Grade", "Reading Score", "Math Score", "School Name", "School ID", "School Type", "School Size", "School Budget"]]

# Calculate the percentage of students who passed math **and** reading (% Overall Passing)
number_of_students_passing_math_and_reading = (len(number_of_students_passing_math_df.loc[number_of_students_passing_math_df["Reading Score"] >= 70]))
percentage_of_students_passing_math_and_reading = ((100/total_number_of_students)*number_of_students_passing_math_and_reading)
```

```python
# Creating additional dataframe for students with a passing math score & a passing reading score (70 or greater in both categories)
number_of_students_passing_math_and_reading_df = number_of_students_passing_math_df.loc[number_of_students_passing_math_df["Reading Score"] >= 70, ["Student Name", "Student ID", "Gender", "Grade", "Reading Score", "Math Score", "School Name", "School ID", "School Type", "School Size", "School Budget"]]

# Format the budget column
number_of_students_passing_math_and_reading_df["School Budget"] = number_of_students_passing_math_and_reading_df["School Budget"].map('${:,.2f}'.format)

#number_of_students_passing_math_and_reading_df
```

```python
# Create a dataframe to hold the above results
district_summary_df = pd.DataFrame({"Total Schools": total_number_of_schools,
                                    "Total Students": total_number_of_students, 
                                    "Total Budget": total_budget, 
                                    "Average Math Score": average_math_score, 
                                    "Average Reading Score": average_reading_score, 
                                    "% Passing Math": percentage_of_students_passing_math, 
                                    "% Passing Reading": percentage_of_students_passing_reading, 
                                    "% Overall Passing": percentage_of_students_passing_math_and_reading},index=[0])

# Format the new dataframe so it is professional and readable
district_summary_df["Total Students"] = district_summary_df["Total Students"].map('{:,}'.format)
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map('${:,.2f}'.format)
district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].map('{:.2f}'.format)
district_summary_df["Average Reading Score"] = district_summary_df["Average Reading Score"].map('{:.2f}'.format)
district_summary_df["% Passing Math"] = district_summary_df["% Passing Math"].map('{:.2f}%'.format)
district_summary_df["% Passing Reading"] = district_summary_df["% Passing Reading"].map('{:.2f}%'.format)
district_summary_df["% Overall Passing"] = district_summary_df["% Overall Passing"].map('{:.2f}%'.format)

district_summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# collect the base data
school_summary_extract = school_data_complete_reformatted.loc[:, ['Student ID', 
                                                                'Reading Score', 
                                                                'Math Score', 
                                                                'School Name', 
                                                                'School Type', 
                                                                'School Size', 
                                                                'School Budget']]

# re-arrange and rename columns
school_summary_extract = school_summary_extract[['School Name', 
                                                'School Type', 
                                                'School Size',
                                                'School Budget',
                                                'Reading Score',
                                                'Math Score']]

school_summary_extract.rename(columns={'School Name': 'school_name'}, inplace=True)
```

```python
# group dataframe by School Name
school_summary_grouped = school_summary_extract.groupby(["school_name"], as_index=False)

# in this instance, the mean will give us the same value for school size and budget as it is a constant value for each school, but it will give us the average math and reading score
school_mean_df = school_summary_grouped.mean()
```

```python
# merge the new school_mean_df df with the schools_complete df to get the school type column. 
school_summary_merge = pd.merge(school_mean_df, school_data, how="left", on=["school_name", "school_name"])

# remove double ups and unnecessary columns
school_summary_merge = school_summary_merge[['school_name', 'Reading Score', 'Math Score', 'type', 'size', 'budget']]
```

```python
# create a column for "Per Student Budget" in the school_summary_merge dataframe
school_summary_merge['Per_Student_Budget'] = school_summary_merge['budget'] / school_summary_merge['size'] 
#school_summary_merge

# Creating dataframe for students passing math & reading - group by school & merge it with the summary df
number_of_students_passing_math_and_reading_df_2 = number_of_students_passing_math_df.loc[number_of_students_passing_math_df["Reading Score"] >= 70, ["Student Name", "School Name"]].groupby(["School Name"], as_index=False)
number_of_students_passing_maths_and_reading_by_school_df = pd.DataFrame(number_of_students_passing_math_and_reading_df_2.size())
number_of_students_passing_maths_and_reading_by_school_df.reset_index(inplace= True)
number_of_students_passing_maths_and_reading_by_school_df.rename(columns={'School Name': 'school_name', 0: 'num of students passing overall'}, inplace=True)
# number_of_students_passing_maths_and_reading_by_school_df
school_summary_merge = pd.merge(school_summary_merge, number_of_students_passing_maths_and_reading_by_school_df, how="left", on=["school_name", "school_name"])

# Creating dataframe for students passing math - group by school & merge it with the summary df
number_of_students_passing_math_df_2 = number_of_students_passing_math_df.groupby(["School Name"], as_index=False)
number_of_students_passing_maths_by_school_df = pd.DataFrame(number_of_students_passing_math_df_2.size())
number_of_students_passing_maths_by_school_df.reset_index(inplace= True)
number_of_students_passing_maths_by_school_df.rename(columns={'School Name': 'school_name', 0: 'num of students passing math'}, inplace=True)
#number_of_students_passing_maths_by_school_df
school_summary_merge = pd.merge(school_summary_merge, number_of_students_passing_maths_by_school_df, how="left", on=["school_name", "school_name"])

# Creating dataframe for students passing reading - group by school & merge it with the summary df
number_of_students_passing_reading_df_2 = number_of_students_passing_reading_df.groupby(["School Name"], as_index=False)
number_of_students_passing_reading_by_school_df = pd.DataFrame(number_of_students_passing_reading_df_2.size())
number_of_students_passing_reading_by_school_df.reset_index(inplace= True)
number_of_students_passing_reading_by_school_df.rename(columns={'School Name': 'school_name', 0: 'num of students passing reading'}, inplace=True)
#number_of_students_passing_reading_by_school_df
school_summary_merge = pd.merge(school_summary_merge, number_of_students_passing_reading_by_school_df, how="left", on=["school_name", "school_name"])

# create a column for "% Passing Math"
school_summary_merge["% Passing Math"] = (100 / school_summary_merge['size']) * school_summary_merge['num of students passing math'] 

# create a column for "% Passing Reading"
school_summary_merge["% Passing Reading"] = (100 / school_summary_merge['size']) * school_summary_merge['num of students passing reading'] 

# create a column for "% Passing Overall"
school_summary_merge["% Passing Overall"] = (100 / school_summary_merge['size']) * school_summary_merge['num of students passing overall'] 
```

```python
# create the School Summary dataframe from the values in school_summary_merge dataframe
school_summary_df = school_summary_merge[['school_name',
                                            'type',
                                            'size',
                                            'budget',	
                                            'Per_Student_Budget',
                                            'Math Score',
                                            'Reading Score',
                                            '% Passing Math',
                                            '% Passing Reading',
                                            '% Passing Overall']]

# Rename columns
school_summary_df.rename(columns={'school_name': 'School Name',
                                    'Reading Score': 'Average Reading Score',
                                    'Math Score': 'Average Math Score',
                                    'type': 'School Type',
                                    'size': 'Total Students',
                                    'budget': 'Total School Budget'	,
                                    'Per_Student_Budget': 'Per Student Budget',
                                    '% Passing Overall': '% Overall Passing'}, inplace=True)

# Format the new dataframe values
school_summary_df["Total School Budget"] = school_summary_df["Total School Budget"].map('${:,.2f}'.format)
school_summary_df["Per Student Budget"] = school_summary_df["Per Student Budget"].map('${:,.2f}'.format)

school_summary_df.round(2)
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# collect the base data
scores_by_school_spend_df = school_summary_merge
#scores_by_school_spend_df.head()

# Figure out the minimum and maximum budget per school
print(scores_by_school_spend_df["Per_Student_Budget"].max())
print(scores_by_school_spend_df["Per_Student_Budget"].min())
```

```python
# Create bins in which to place values based upon budget per school
bins = [0, 585, 630, 645, 676]

# Create labels for these bins
group_labels = ["< $584", "$585-629", "$630-644", "$645-675"]

# Slice the data and place it into bins
pd.cut(scores_by_school_spend_df["Per_Student_Budget"], bins, labels=group_labels).head()
```

```python
# Place the data series into a new column inside of the DataFrame
scores_by_school_spend_df["Spending Ranges (Per Student)"] = pd.cut(scores_by_school_spend_df["Per_Student_Budget"], bins, labels=group_labels)
# scores_by_school_spend_df.head()
```

```python
# Create a GroupBy object based upon "Spending Ranges (Per Student)"
spending_ranges = scores_by_school_spend_df.groupby("Spending Ranges (Per Student)")

# Get the average of each column within the GroupBy object
spending_ranges[["Math Score", "Reading Score", "% Passing Math", "% Passing Reading", "% Passing Overall"]].mean()
```

```python
# make the grouped data into it's own dataframe
spending_ranges_df = pd.DataFrame(spending_ranges[["Math Score", "Reading Score", "% Passing Math", "% Passing Reading", "% Passing Overall"]].mean())

# Rename the columns
spending_ranges_df.rename(columns={'Math Score': 'Average Math Score', 'Reading Score': 'Average Reading Score'}, inplace=True)

# Round the numbers to 2 decimal places
spending_ranges_df.round(2)
```
